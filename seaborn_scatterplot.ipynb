```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = sns.load_dataset('tips')
data.head(5)
```

```python
plt.style.use('fivethirtyeight')
```

Scatterplot b/w tips and total_bill

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data) 
# relplot(figure level) is relational plot which can plot both line and the scatterplot.
```

The Hue Parameter


```python
sns.relplot(x='total_bill',y='tip',hue='smoker',kind='scatter',data=data) 
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',kind='scatter',data=data) 
```

categorical scatterplot


```python
plt.style.use('fivethirtyeight')
```

```python
sns.catplot(x='day',y='tip',kind='strip',data=data)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=data)
```

```python
sns.catplot(x='day',y='tip',kind='swarm',hue='sex',data=data)
```
