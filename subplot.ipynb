```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data (replace this with your own data)
data = {'Category': [ 'B', 'A', 'C', 'B', 'A', 'C', 'C', 'B', 'A']}
df = pd.DataFrame(data)

# Get the order of x-axis labels based on counts
order = df['Category'].value_counts().index

# Create a countplot with sorted xtick labels
sns.countplot(x='Category', data=df, order=order)

# Set x-axis label
plt.xlabel('Category')

# Set y-axis label
plt.ylabel('Count')

# Set plot title
plt.title('Countplot with Sorted Xtick Labels')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame with a datetime column named 'date'
# Create a sample DataFrame for demonstration
date_rng = pd.date_range(start='2010-01-01', end='2020-12-31', freq='D')
df = pd.DataFrame(date_rng, columns=['date'])

# Create a new column for years and months
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# Create subplots
fig, axes = plt.subplots(10, 1, figsize=(20, 20), sharex=True)

# Loop through each year and plot countplot for each month
for i, year in enumerate(range(2010, 2021)):
    df_year = df[df['year'] == year]
    ax = sns.countplot(x='month', data=df_year, ax=axes[i])
    ax.set_title(f'Year {year}')
    ax.set_xlabel('Month')
    ax.set_ylabel('Count')
# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Plot all lines in a single line without using subplots
plt.plot(x, y1, label='Sin(x)', color='blue')
plt.plot(x, y2, label='Cos(x)', color='red')
plt.plot(x, y3, label='Tan(x)', color='green')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.random.rand(100) * 5  # Sample data for a scatter plot

# Create the first plot (line plot)
fig, ax1 = plt.subplots()

ax1.plot(x, y1, label='Sin(x)', color='blue')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Sin(x)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create the second plot (scatter plot) on the same x-axis
ax2 = ax1.twinx()
ax2.scatter(x, y2, label='Random Scatter', color='red')
ax2.set_ylabel('Random Scatter', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add a common title
plt.title('Two Different Plots on the Same Horizontal Line')

# Show the plot
plt.show()
```
