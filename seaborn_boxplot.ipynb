Boxplot
A boxplot is a standarized way of displaying the distribution of data based on the five number of summery("minimum",first quartile(Q1),median,third quartile(Q3),and "maximum").It can tell about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped and how your data is skewed.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = sns.load_dataset('tips')
data.head(5)
```

```python
plt.style.use('fivethirtyeight')
```

```python
sns.catplot(x='day',y='tip',kind='box',data=data)
```

```python
sns.catplot(x='day',y='tip',kind='box',hue='sex',data=data)
```

Barplot

A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',data=data)
# the black bars are the arraybars that are showing confidential interval to calculate the mean for the population.
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',hue='sex',data=data)
```

```python
# the default behaviour of blackbars is to calculate the mean of numerical value bt we can change it using estimater.
#estimator=np.median/np.var/np.mean
sns.catplot(x='smoker',y='total_bill',kind='bar',hue='sex',estimator=np.median,data=data)
```

```python
sns.catplot(x='smoker',y='total_bill',kind='bar',hue='sex',estimator=np.var,data=data)
```

Countplot

Countplot is special case of  barplot bt only difference is countplot counts the categories

```python
sns.catplot(x='smoker',kind='count',hue='sex',estimator=np.var,data=data)
```

HeatMap

A heatmap is graphical representation of data where the individual values contained in a matrix are represented as color.(drwan b/w categorical data and sometime discrete numerical data).

```python
flights = sns.load_dataset('flights')
flights.head(5)
```

```python
dt = flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
sns.heatmap(dt)
```

```python
sns.heatmap(dt,linewidths=0.9,annot=True,fmt='d')
```

```python
sns.heatmap(dt,linewidths=0.9,annot=True,fmt='d',cmap='summer')   # cmap=color , fmt is value in the blocks
```

Cluster Map

```python
sns.clustermap(dt)
```

```python
sns.clustermap(dt,metric='correlation')
```

```python
sns.clustermap(dt,z_score=0,annot=True,row_cluster=False,metric='correlation')
```

```python
sns.clustermap(dt,z_score=0,annot=True,col_cluster=False,metric='correlation')
```

Pairplot

Pairplot are used to plot mulitple bivariate distribution in a dataset . This create a matrix of axes and shows the relationship for pair of columns in a dataframe. By default, it  also draws the univariate distribution of each variable on the diagonal axes.

```python
iris = sns.load_dataset('iris')
iris.head(5)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue ='species')
```
