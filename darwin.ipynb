```python
for i in range(1,5):
    data = pd.read_csv("darwin.csv")
    data = data.sample(frac=1).reset_index(drop=True)
    y = data.iloc[:,[-1]]
    x= data.iloc[:,1:451]
    ts = 0.10*i
    from sklearn.preprocessing import OneHotEncoder
    encoder = OneHotEncoder(drop='first',sparse=False)
    y_new = encoder.fit_transform(y)
    y = pd.DataFrame(y_new,columns = y.columns)
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    x = scaler.fit_transform(x)
    from sklearn.model_selection import train_test_split
    x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0,test_size=ts)
    from sklearn.neighbors import KNeighborsClassifier
    classifier = KNeighborsClassifier(n_neighbors=10) # n_neighbors = number nearest neighbors , p = Euclidean metric
    classifier.fit(x_train,y_train)                     # Euclidean distance = distance between the best fit line and the nearest points.
    y_pred = classifier.predict(x_test)
    import sklearn.metrics as sm
    print("accuracy : ", (sm.r2_score(y_test,y_pred)))
    from sklearn.metrics import confusion_matrix
    gv = confusion_matrix(y_test,y_pred)
    print(gv)
    
```

```python
for i in range(1,5):
    data = pd.read_csv("darwin.csv")
    data = data.sample(frac=1).reset_index(drop=True)
    y = data.iloc[:,[-1]]
    x= data.iloc[:,1:451]
    ts = 0.10*i
    from sklearn.preprocessing import OneHotEncoder
    encoder = OneHotEncoder(drop='first',sparse=False)
    y_new = encoder.fit_transform(y)
    y = pd.DataFrame(y_new,columns = y.columns)
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    x = scaler.fit_transform(x)
    from sklearn.model_selection import train_test_split
    x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0,test_size=ts)
    from sklearn.linear_model import Lasso
    lasso_regressor = Lasso(alpha=0.3)
    lasso_regressor.fit(x_train,y_train)
    y_pred = lasso_regressor.predict(x_test)
    import sklearn.metrics as sm
    print("accuracy : ", (sm.r2_score(y_test,y_pred))*100)
    
```
